### 2.2 signals and systems

Textbooks:  [Systems, Signals, and Transforms](https://alliance-primo.hosted.exlibrisgroup.com/primo-explore/fulldisplay?docid=CP71259169820001451&context=L&vid=UW&lang=en_US) by Phillips, Parr, and Riskin; [Signals and Systems](https://alliance-primo.hosted.exlibrisgroup.com/primo-explore/fulldisplay?docid=CP71145530020001451&context=L&vid=UW&lang=en_US) by Oppenheim, Willsky, and Hamid.

In the following problems, let 
$S$ be a single-input/single-output (SISO) linear time-invariant (LTI) system with time-domain ***impulse response*** $h:\mathbb{R}\rightarrow\mathbb{R}$,
let
$u:\mathbb{R}\rightarrow\mathbb{R}$ be a time-domain ***input*** to the system,
and let
$y:\mathbb{R}\rightarrow\mathbb{R}$ be the time-domain ***output*** corresponding to input $u$.

a. Give a time-domain expression for $y$, that is, write $y(t)$ in terms of $h$ and $u$; be explicit about any notation you introduce.

---

**Solution:** In time-domain, the system can be expressed as\
$y(t)=u(t)\otimes h(t) = \int_{-\infty}^\infty u(\tau)h(t-\tau) d\tau, \tau\in t$\
where '$\otimes$' means convolution.

---

$y(t) = (h \otimes x)(t) = \int_{-\infty}^{\infty} h(t-\tau)x(\tau)d\tau$ where $\otimes$ is the convolution operator given implicitely by the RHS.

Let 
$H = \mathcal{F}h$, 
$Y = \mathcal{F}y$, 
$U = \mathcal{F}u$
denote the Fourier transforms of the time-domain impulse response, input, and output, respectively.

b. Give a frequency-domain expression for $Y(\omega)$, that is, write $Y(\omega)$ in terms of $H(\omega)$ and $U(\omega)$.

---

**Solution:** In time-domain, the system can be expressed as\
$Y(j\omega) = H(j\omega)U(j\omega)$

***Notice:***  your expression for $y(t)$ in (a.) depends on the value of $h$ and $u$ at ***all times*** (not just the time $t$), whereas your expression for $Y(\omega)$ in (b.) depends only on the value of $H$ and $U$ at frequency $\omega$.
This illustrates the key strength of frequency-domain analysis -- convolution in the time domain turns into multiplication in the frequency domain.

---

c. Give the general expression for the Fourier transformation $\mathcal{F}$, that is, write $H(\omega)$ in terms of $h$.

---

**Solution:** \
$H(j\omega) = \int_{-\infty}^{\infty} h(t) e^{-j\omega t} dt$
Ref: https://en.wikipedia.org/wiki/Fourier_transform

---

d. Give the general expression for the ***inverse*** Fourier transformation $\mathcal{F}^{-1}$, that is, write $h(t)$ in terms of $H$.

---

**Solution:** According to inverse Fourier transform,

$h(t) = \int_{-\infty}^{\infty} H(j\omega)e^{j\omega t} d\omega$

***Notice:*** since an integral is an infinite sum, you can infer from the expressions in (c.) and (d.) that 
the time-domain signal $h$ can be represented as an infinite linear combination of frequency-domain exponentials,
and similarly that 
the frequency-domain signal $H$ can be represented as an infinite linear combination of time-domain exponentials.

### 2.3 linear algebra 

Textbooks:  [Introduction to Applied Linear Algebra](http://vmls-book.stanford.edu/) by Boyd and Vandenberghe; [Linear Algebra](http://joshua.smcvt.edu/linearalgebra/) by Hefferon.

In the following problems, let 
$A = \left[\begin{array}{} A_{11} & A_{12} \\ 0 & A_{22} \end{array}\right]\in\mathbb{R}^{2\times 2}$ 
be a matrix with ***shape*** $2\times 2$,
$B = \left[\begin{array}{} B_{11} \\ B_{21} \end{array}\right]\in\mathbb{R}^{2\times 1}$
be a matrix with shape $2\times 1$,
and
$x = \left[\begin{array}{} x_{1} \\ x_{2} \end{array}\right]\in\mathbb{R}^{2}$
be a vector with ***length*** $2$. 

a. Determine the eigenvalues of $A$.

---
**SOLUTION**
If $\lambda$ is an eigenvalue of $A$, and $x \neq 0$ is the corresponding eigenvector, then $(\lambda, x)$ must satisfy

$$ (A-\lambda I)x = 0 $$

since $x \neq 0$, the equation above only has a solution if the determinant $|A-\lambda I| = 0$. Therefore, the eigenvalues are roots of the *characteristic polynomial*

$$ (A_{11}-\lambda)(A_{22}-\lambda) = 0$$

for which there are clearly two solutions $\lambda_1, \, \lambda_2$

$$ \fbox{ $ \lambda_1 = A_{11}, \, \lambda_2 = A_{22}$ }$$

---

b. Compute the determinant of $A$.

---
**SOLUTION**
The general formula for the determinant of a $2 \times 2$ matrix $A$ is 

$$ |A| = A_{11} A_{22} - A_{12}A_{21}$$

where $A_{ij}$ is the entry of $A$ in row $i$ and column $j$. So, if $A_{21}=0$, we have

$$ |A| = A_{11} A_{22} $$

---

c. Compute $Ax$ and $x^\top A^\top$ using matrix multiplication, where $M^\top$ denotes the ***transpose*** of matrix $M$.

---
**SOLUTION**

Given two matrices (or vectors) with compatible dimension $A \in \mathbb{R}^{n,m}$ and $B \in \mathbb{R}^{m,p}$ the general formula for matrix multiplication $AB$ is 

$$ (AB)_{ij} = \sum_{k=1}^m A_{ik}B_{kj}$$

Hence if  $y = Ax$ then the entries of $y$ are given by 

$$ 
y = \left[ \begin{array}{c} A_{11}x_1 + A_{12}x_2 \\ A_{22}x_2\end{array}\right]
$$

---

d. Why can't you compute $x A$ or $A^\top x^\top$ using matrix multiplication?

---
**Solution**

The usual rules for matrix multiplication require that the *inner dimension* of the operands agree. In other words, to calculate $AB$, we require that the number of columns in $A$ equals the number of rows in $B$. If $x$ is a vector in $\mathbb{R}^n$, then we treat $x$ as an $(n \times 1)$ matrix for the purpose of calculating $Ax$.

In this way, $Ax^\top$ does not make sense, because we would be trying to multiply a $(n \times n)$ matrix by a $(1 \times n)$ matrix, whose inner dimensions do not agree. Theoretically, $x^\top$ is not a vector but it is a *dual vector*, and hence $x^\top$ is not in the domain of left-multiplication by $A$.

The exact same argument can be made used to show that $xA$ is not meaningful. Practically speaking, their inner dimensions do not agree. Theoretically , right-multiplication by a $A$ is a linear map between *dual vector spaces*, but $x$ is a vector, and so is not in the domain of right-multiplication by $A$.

---

e. Compute $B + x$; is the result a matrix or vector?

---
**Solution**
While it possible to *define* a kind of addition to calculate $x+B$, you should pause first. Why? Because $B$ is a matrix, while $x$ is a vector, and addition between matrices and vectors of any shape is not defined. Should $B+x$ be a vector, or a matrix? Actually, we cannot say without making an (arbitrary) choice!

To expand on this point a little further, we want to distinguish $\mathbb{R}^{n \times 1}$ from $\mathbb{R}^n$. Both are naturally vectorspaces, and each have their own addition operation,

$$ +_{n \times 1} : \mathbb{R}^{n \times 1} \times \mathbb{R}^{n \times 1} \to \mathbb{R}^{n \times 1}$$

is the addition on $\mathbb{R}^{n \times 1}$, and 

$$ +_n : \mathbb{R}^n \times \mathbb{R}^n \to \mathbb{R}^n $$

is the addition on $\mathbb{R}^n$. However, without making a choice about whether the result of $B+x$ is a matrix or a vector, we cannot define an addition that takes an element of $\mathbb{R}^{n \times 1}$ and adds it to an element of $\mathbb{R}^n$.

*Note: in numpy, if you try to naively add a 1-column ndarray (a matrix) to an array of the same size (vector), you will probably not get the result you were expecting. *
