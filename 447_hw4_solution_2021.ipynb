# 1. step response of a linear system

Consider the scalar DE
$$T \dot{y} + y = u.$$

**Assume the input is constant ($u(t) = u_0$).**

a. Express $y(t)$ in terms of $t$, $T$, $y(0)$, $u_0$.

**Solution:**\
*Homogenous solution:*    

Characteristic equation: $s + 1/T = 0$    
$\Leftrightarrow s = -1/T$    
The homogenous solution is $y_h(t) = c_1e^{-t/T}$    

*Particular solution:*    

$\dot{y} + \frac{y}{T} = \frac{u_0}{T}$    

Therefore, particular solution is:    

$y_p(t) = u_0$

Therefore, $y(t) =  c_1e^{-t/T} + u_0$. Because $y(0) = c_1+u_0$, $y(t) = (y(0)-u_0)e^{-t/T} + u_0$.

-------------------------------------

b. Show that the *steady-state value* equals the constant input: $y_{ss} = u_0$.   

**Solution:**
As t increases, 

$y_{ss} = \lim_{t\rightarrow\infty} y(t) = (y(0)-u_0) \times 0 + u_0  = u_0$


--------------------------

**Now assume the initial condition is zero, $y(0) = 0$; in this case the output is termed the *step response*.**

c. Show that the *rise time* of the step response (i.e. the time required for the output to go from $10\%$ to $90\%$ of its steady-state value) is approximately $2T$.

**Solution:**\
If $y(0) = 0$, \
$y(t)=(1-e^{-t/T})u_0$\
If the output is 10% steady-state value,\
$e^{-t_1/T} = 0.9, \Rightarrow t_1=-ln(0.9)T=0.105T$\
If the output is 90% steady-state value,\
$e^{-t_2/T} = 0.1, \Rightarrow t_2=-ln(0.1)T=2.303T$\
Thus, it takes approximately $2T$ for the output go from  $10\%$  to  $90\%$.


-----------------------------
d. Show that the *settling time* of the step response (i.e. the time required for the output to reach and stay within $2\%$ of its steady-state value) is approximately $4T$.

**Solution:**\
Similarly as 1c., $y(t)=(1-e^{-t/T})u_0$,\
When output reaches $2\%$ of the steady-state value,\
$e^{-t_3/T} = 0.02, \Rightarrow t_2=-ln(0.02)T=3.912T$

**Now recall the definition of the *matrix exponential*:**

$$ \forall A \in \mathbb{R}^{n\times n} : e^A = \sum_{k=0}^\infty \frac{1}{k!} A^k. $$

**In the remainder of this problem, you'll use this definition directly to establish useful facts.**

(b.) Show that every square matrix commutes with its matrix exponential, i.e. $A\, e^A = e^AA$ for all $A\in\mathbb{R}^{n\times n}$.

**Solution:**

Starting from the L.H.S:   
$A\, e^A = A\, \sum_{k=0}^\infty \frac{1}{k!} A^k$ 

For square matrices, the matrix exponential can be re-written as:    
$A\, e^A  =  A(I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...) $    
$\Leftrightarrow  A + AA+\frac{AA^2}{2!}+ \frac{AA^3}{3!}...$  
$\Leftrightarrow  A + A^2+\frac{A^3}{2!}+ \frac{A^4}{3!}...$     

By factoring out A to the right,    
$\Leftrightarrow  A\, e^A = (I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...)A $   
$\Leftrightarrow A\, e^A =   e^AA$   

Therefore, square matrices commute with their matrix exponentials.

(c.) Given an invertible matrix $T\in\mathbb{R}^{n\times n}$, show that $e^{T A T^{-1}} = T e^A T^{-1}$.

**Solution:**

$e^{T A T^{-1}} = T e^A T^{-1}$   
$e^A = \sum_{k=0}^\infty \frac{1}{k!} A^k$     
$\Leftrightarrow e^{T^{-1} AT} = I + T^{-1} AT+\frac{(T^{-1}AT)^2}{2!}+ \frac{(T^{-1}AT)^3}{3!}...$     

By examining the terms in the infinite sum:    

$(T^{-1}AT)^2 = T^{-1}ATT^{-1}AT$ 
$\Leftrightarrow (T^{-1}AT)^2 = T^{-1}AIAT$ (Since $T^{-1}T = I$)    
$\Leftrightarrow (T^{-1}AT)^2 = T^{-1}A^2T$

Similarly, for each of the product terms, we get $(T^{-1}AT)^n = T^{-1}A^nT$.

Therefore,     
$\Leftrightarrow e^{T^{-1} AT} = I + T^{-1} AT+\frac{T^{-1}A^2T}{2!}+ \frac{T^{-1}A^3T}{3!}...$
$\Leftrightarrow e^{T^{-1} AT} = T^{-1}IT + T^{-1} AT+\frac{T^{-1}A^2T}{2!}+ \frac{T^{-1}A^3T}{3!}...$
$\Leftrightarrow e^{T^{-1} AT} = T^{-1}(I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...)T$
$\Leftrightarrow e^{T^{-1} AT} = T^{-1}e^AT$

(d.) Create a root locus plot for the closed-loop system as parameter $k_P$ varies.



```
import numpy as np
import pylab as plt

q = 0.05
dq = 0
i = 1.72

#Choosing values of k for annotation

kps = np.linspace(-200,2000,70)

lambda1 = np.array([])
lambda2 = np.array([])
lambda3 = np.array([])
for kp in kps:
  A = np.array([[0, 1, 0],
              [-0.0166*(i**2)*(q**(-3)), 0, 0.0166*i*(q**(-2))],
              [-4*10**(-4)*i*dq*q**(-3)-kp, 2*10**(-4)*i*q**(-2), 2*10**(-4)*dq*q**(-2)-30]])

  eigenvalue, _ = np.linalg.eig(A)

  lambda1 = np.append(lambda1, eigenvalue[0])
  lambda2 = np.append(lambda2, eigenvalue[1])
  lambda3 = np.append(lambda3, eigenvalue[2])

print(np.shape(lambda1))
# confirm the non-real eigenvalues are complex-conjugates
assert np.allclose(lambda1[np.abs(lambda1.imag) > 0.],
                   np.conj(lambda2[np.abs(lambda1.imag) > 0]))

#Axis labels
plt.title('locus of eigenvalues as feedback parameter $k$ varies')
plt.xlabel(r'real part of eigenvalue ($\operatorname{Re}\ \lambda$)')
plt.ylabel(r'imaginary part of eigenvalue ($\operatorname{Im}\ \lambda$)')

#Plotting values
plt.scatter(lambda1.real,lambda1.imag,color='b')
plt.scatter(lambda2.real,lambda2.imag,color = 'r')
plt.scatter(lambda3.real,lambda3.imag,color = 'g')
# we'll only annotate every 10 of the computed eigenvalues
for k,lam1,lam2 in list(zip(kps,lambda1,lambda2))[::10]:
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam1.real,lam1.imag))
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam2.real,lam2.imag))
#x- and y- limits and drawing axes
ylim = plt.ylim()
plt.plot(np.zeros(2),ylim,'k--')
xlim = plt.xlim()
plt.plot(xlim,np.zeros(2),'k--')
plt.xlim(xlim); plt.ylim(ylim);
```
